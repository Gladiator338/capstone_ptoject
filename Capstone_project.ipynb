```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt


%matplotlib inline

data = pd.read_csv("Indian_Liver_Patient_Dataset.csv")
data.info()
data.head(n=10)
```

```python
n_records = data.shape[0]

n_female =  sum( 1 for _,gender in data.iterrows() if gender['Gender'] == "Female")

n_male =  sum( 1 for _,gender in data.iterrows() if gender['Gender'] == 'Male')

n_percent_females = n_female/n_records*100

n_percent_males = n_male/n_records*100

print("{:.4f}% of females and {:.4f}% of males among {} records ".format(n_percent_females,n_percent_males,n_records))
```

```python
data['Gender'] = data['Gender'].apply(lambda x : 1 if x =='Male' else 0)
data.head(n=10)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.grid_search import GridSearchCV
model=RandomForestClassifier()
params={'criterion':['gini','entropy'],'n_estimators':[100,10]}
disease= data['sector']
new_data=data.drop('sector',axis=1)
X_train,X_test,y_train,y_test=train_test_split(new_data,disease,random_state=0)
clf1=GridSearchCV(estimator=model,param_grid=params,scoring='accuracy',cv=10)
clf1.fit(X_train,y_train)
prediction=clf1.predict(X_test)
from sklearn.metrics import accuracy_score
accuracy=accuracy_score(y_test,prediction)
print(accuracy)
```

```python
from sklearn.linear_model import LogisticRegression
clf2 = LogisticRegression(random_state=0)
clf2.fit(X_train,y_train)
prediction = clf2.predict(X_test)
print(prediction)
from sklearn.metrics import accuracy_score
accuracy=accuracy_score(y_test,prediction)
print(accuracy)
```

```python
from sklearn import svm
from sklearn.model_selection import train_test_split
disease= data['sector']
new_data=data.drop('sector',axis=1)
X_train,X_test,y_train,y_test=train_test_split(new_data,disease,test_size=0.2)
clf3=svm.SVC(kernel='linear',C=1)
clf3.fit(X_train,y_train)
prediction=clf3.predict(X_test)
from sklearn.metrics import accuracy_score
accuracy=accuracy_score(y_test,prediction)
print(accuracy)
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split
disease= data['sector']
new_data=data.drop('sector',axis=1)
X_train,X_test,y_train,y_test=train_test_split(new_data,disease,test_size=0.3)
clf4= KNeighborsClassifier(n_neighbors=4,weights='uniform')
clf4.fit(X_train,y_train)
prediction=clf4.predict(X_test)
from sklearn.metrics import accuracy_score
accuracy=accuracy_score(y_test,prediction)
print(accuracy)
```
